## Zufallszahlen
In Python lassen sich relativ schnell Zahlen in einem Berech generieren. Zum Beispiel mit einer Schleife oder kürzer mit einer List Comprehension

```python
x=[]
for n in range(-100,101,1):
    x.append(n/10)
## List Comprehension
x = [n/10 for n in range(-100,101,1)]
```

Für Zufallszahlen kann man die Random Funktion benutzen

```python
import random

print(random.random())      #Zufallszahl zwischen 0 und 0.9999
print(random.random()*9+1)  #Oder mit angepasstem Intervall
print(random.randint(1,6))     #Zufallszahl Ganzzahl mit Min und Max

print([random.randint(1,10) for n in range(0,11)])      #Oder aber auch zum generieren von 
                                                        #Zufallszahlen in einer Liste
```

Numpy kann auch Zufallszahlen generien

```python
import numpy as np

#np.random.seed(0)       # Funktion mit der wir immer die gleichen Zufallszahlen generieren können

print(np.random.random(10))     #Generieren eines Arrays mit 10 Werten
print(np.random.random(10)*9+1) #Oder angepasstes Intervall
print(np.random.randint(1,6,10))   #Zufallszahl mit Min und Max und anzahl der Elemente
```

## Matplotlib
Matplotlib ist eine visualisierungs Bibliothek in Python, mit der zum Beispiel Graphen gezeichnet werden können.

```python
from matplotlib import pyplot as plt

x=[n for n in range(-5,5,1)]
y=[2*n +3 for n in x]                         #Geradengleichung

plt.plot(x,y)                                 #Werte für die Koordinaten x-Achse und y-Achse
plt.show                                      #Graph Anzeigen
```

Die Diagramme lassen sich natürlich noch verschöner mit Achsenbeschriftung, Titel, usw.

```python
x=[n for n in range(-5,5,1)]
y=[-2*n -3 for n in x]                         

plt.plot(x,y)
plt.xlabel("X-Werte")                       #Beschriftung X-Achse
plt.ylabel("Y-Werte")                       #Beschriftung y-Achse
plt.title("Funktion f(x)=-2x-3")            #Beschriftung Titel
plt.grid()                                  #Anzeige Hilfslinien
plt.show 
```

Möchte man Graphen mit einander verlgeichen, geht dies auch im gleichen Diagramm durch sogenannte Subplots.

```python
from matplotlib import pyplot as plt

x=[n for n in range(-5,5,1)]
y=[2*n +3 for n in x]                         

plt.subplot(1,2,1)              # plt.subplot(anzahl zeilen, anzahl spalten, nummer des bildes)
plt.plot(x,y)                                 
plt.show  



x=[n for n in range(-5,5,1)]
y=[-2*n -3 for n in x]                         
plt.subplot(1,2,2)
plt.plot(x,y)
plt.xlabel("X-Werte")                       
plt.ylabel("Y-Werte")                       
plt.title("Funktion f(x)=-2x-3")            
plt.grid()                                  
plt.show
```

Mit Scatter lassen sich Steuungen von Datenpunkten visualisieren. So kann man schnell Ausreißer in einem Datensatz erkennen.

```python

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
#plt.grid()
plt.show()
```

Ebenso gibt es die Möglichkeit sich Histogramme anzeigen zu lassen

```python

x = np.random.normal(170, 10, 250)              #random.normal liefert uns eine Normalverteilte 
                                                #Zufallsvariable, dazu kommen wir in der nächsten 
                                                #Stunde

plt.hist(x)
plt.show() 
```
