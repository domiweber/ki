### Zufallszahlen
Der Computer soll einen Würfel ersetzen.<br>
Erstellen Sie eine Liste mit 20 Zufallszahlen geben Sie die Liste als Punkte-Graph aus (plt.scatter()). Für die x-Werte können Sie die Zahlen von 1 bis 20 verwenden.

```python
from matplotlib import pyplot as plt
import numpy as np
import random

x = []

for i in range(1,20,1):
    x.append(random.random()*20)
y = []

for i in range(1,20,1):
    y.append(random.random()*20)
plt.scatter(x,y)            # plt.subplot(anzahl zeilen, anzahl spalten, nummer des bildes)   
plt.show   
```
